```python
set_name = 'Training Set/' 
# dataframe = pd.DataFrame()
corpus = ""
for i in os.listdir(set_name):
    parsed = parser.from_file(set_name+i)
    parsed['content'] = parsed['content'].replace('\n',' ')
    parsed['content'] = parsed['content'].replace('\t','')
#     cleaned_content = re.sub('[(]{1}[A-z\s]+[)]{1}',' ',parsed['content'])
#     cleaned_content = lemmatizer.stem(cleaned_content)
    corpus += parsed['content']
counts = collections.Counter(corpus.split())
most_common_words= [word.lower() for word, word_count in counts.most_common(50)]
most_common_words.remove('on') #on,and,month,by
most_common_words.remove('and')
most_common_words.remove('month')
most_common_words.remove('by')
most_common_words.remove('rent')
most_common_words.remove('pay')
most_common_words.remove('agrees')
most_common_words.append('rental')
most_common_words.append('agreement')
most_common_words.append('house')

```

```python
dataframe = pd.DataFrame()
total_match = 0
name_match = 0
for i in os.listdir(set_name):
    parsed = parser.from_file(set_name+i)
    parsed['content'] = parsed['content'].replace('\n',' ')
    parsed['content'] = parsed['content'].replace('\t','')
    file_name = os.path.splitext(os.path.splitext(i)[0])[0]
#     print(file_name)
    if file_name in training_set['File Name'].values:
        total_match += 1
#         print('Aggrement Value:',training_set[training_set['File Name']==file_name]['Aggrement Value'].values[0])
#         print('Aggrement Start Date:',training_set[training_set['File Name']==file_name]['Aggrement Start Date'].values[0])
#         print('Aggrement End Date:',training_set[training_set['File Name']==file_name]['Aggrement End Date'].values[0])
#         print('Renewal Notice (Days):',training_set[training_set['File Name']==file_name]['Renewal Notice (Days)'].values[0])
#         print('Party One:',training_set[training_set['File Name']==file_name]['Party One'].values[0])
#         print('Party Two:',training_set[training_set['File Name']==file_name]['Party Two'].values[0])
#         print(training_set[training_set['File Name']==file_name]['Party Two'].values[0])
#         print(i,parsed['content'])
#         print('\n\n',parsed['content'].lower().split('follows')[0])
#         print(' '.join(nltk.sent_tokenize(parsed['content'])))
#         corpus += parsed['content']
#         cleaned_para = ' '.join(nltk.sent_tokenize(parsed['content'])[:4])
#         cleaned_para = cleaned_para.lower()
#         cleaned_para = cleaned_para.translate(str.maketrans('','',string.punctuation))
#         cleaned_para = re.sub('MR|Mr|MRS|Mrs','',cleaned_para)
#         cleaned_para = word_tokenize(cleaned_para)
#         print('cleaned para:',cleaned_para)
#         if training_set[training_set['File Name']==file_name]['Party One'].values[0].lower().strip().translate(str.maketrans('', '', string.punctuation)) in cleaned_para:
#             print('Yes')
#         cleaned_content = re.sub('MRS[\s].|MRS|Mrs[\s].|Mrs|MR[\s].|MR|Mr[\s].|Mr|RS[\s].|RS|Rs[\s].|Rs',' ',parsed['content'])

#             cleaned_content = re.sub('[A-z]+[!"#$%&\'()*+-,.:;<=>?@[\\]^_`{|}~]',' ',cleaned_content)
#             cleaned_content = re.sub('[!"#$%&\'()*+-,.:;<=>?@[\\]^_`{|}~]',' ',cleaned_content)
#             cleaned_content = re.sub('[A-z][!"#$%&\'()*+-,.:;<=>?@[\\]^_`{|}~]',' ',cleaned_content)
#         cleaned_content = re.sub('rs.','',cleaned_content)
        cleaned_content = word_tokenize(parsed['content'])
        cleaned_content = [w for w in cleaned_content if not w.lower() in most_common_words] 
        cleaned_content = " ".join(cleaned_content)
#         print(bcolors.OKBLUE,'cleaned_content:',bcolors.ENDC,cleaned_content)
#         cleaned_content = re.sub('Rs.','',cleaned_content)
        cleaned_content = re.sub('[(]{1}[A-z\s]+[)]{1}',' ',cleaned_content)
        cleaned_content = re.sub('[A-z]{1}/[A-z]{1}',' ',cleaned_content)
        cleaned_content = re.sub('[!"#$%&\'()*+-,.:;<=>?@[\\]^_`{|}~]',' ',cleaned_content)
#         cleaned_content = re.sub('\s+[0-9]\s+',' ',cleaned_content)
        encoded_string = cleaned_content.encode("ascii", "ignore")
        cleaned_content = encoded_string.decode()
        cleaned_content = ' '.join(cleaned_content.split())
        #Tokenizing here because later we are subsitituting . which is causing wrong sentence segmentation
#         cleaned_content = nltk.sent_tokenize(cleaned_content)
        print(bcolors.BOLD,'cleaned content:',bcolors.ENDC,cleaned_content)
#         cleaned_content = cleaned_content.lower()
        doc = nlp(cleaned_content)
        cleaned_content = list(doc.sents)
        print(bcolors.BOLD,'cleaned content:',bcolors.ENDC,cleaned_content)
#         for content in cleaned_content:
#             print('content:',str(content))
#             docs=nlp(str(content))
#             pos_tags = [w.pos_ for w in docs]
#             print('pos:',pos)
#         cleaned_content_1=[]
#         for content in cleaned_content:
#             content = re.sub('[,.]',' ',content)
#             content = " ".join(content.split())
#             encoded_string = content.encode("ascii", "ignore")
#             content = encoded_string.decode()
#             if len(content) > 2:
#                 cleaned_content_1.append(content)
#         cleaned_content = list(zip(cleaned_content,pos_tags))
        print(bcolors.OKGREEN,'cleaned content_1:',bcolors.ENDC,cleaned_content)
        for index,elements in enumerate(cleaned_content):
            dataframe = dataframe.append(pd.DataFrame({'filename':file_name,'sentence':'Sentence '+str(index),'content':elements}),ignore_index=True)   
        name_match += 1
        print('\n\n***************************\n\n')
```
