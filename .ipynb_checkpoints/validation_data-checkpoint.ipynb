```python
from __future__ import unicode_literals, print_function

import plac
import random
import warnings
from pathlib import Path
import spacy
from spacy.util import minibatch, compounding


nlp = spacy.blank("en")
# create the built-in pipeline components and add them to the pipeline
# nlp.create_pipe works for built-ins that are registered with spaCy
if "ner" not in nlp.pipe_names:
    ner = nlp.create_pipe("ner")
    nlp.add_pipe(ner, last=True)
# otherwise, get it so we can add labels
else:
    ner = nlp.get_pipe("ner")

# add labels
for _, annotations in train_data:
    for ent in annotations.get("entities"):
        ner.add_label(ent[2])

# get names of other pipes to disable them during training
pipe_exceptions = ["ner", "trf_wordpiecer", "trf_tok2vec"]
other_pipes = [pipe for pipe in nlp.pipe_names if pipe not in pipe_exceptions]
# # only train NER
with nlp.disable_pipes(*other_pipes), warnings.catch_warnings():
    # show warnings for misaligned entity spans once
    warnings.filterwarnings("once", category=UserWarning, module='spacy')

#     # reset and initialize the weights randomly – but only if we're
#     # training a new model
nlp.begin_training()
for itn in range(250):
    random.shuffle(train_data)
    losses = {}
    # batch up the examples using spaCy's minibatch
    batches = minibatch(train_data, size=compounding(4.0, 32.0, 1.001))
    for batch in batches:
        texts, annotations = zip(*batch)
        nlp.update(
            texts,  # batch of texts
            annotations,  # batch of annotations
            drop=0.5,  # dropout - make it harder to memorise data
            losses=losses,
        )
    print("Losses", itn, losses)

# test the trained model
#     for text, _ in train_data:
#         doc = nlp(text)
#         print("Entities", [(ent.text, ent.label_) for ent in doc.ents])
#         print("Tokens", [(t.text, t.ent_type_, t.ent_iob) for t in doc])

# save model to output directory
# if output_dir is not None:
#     output_dir = Path(output_dir)
#     if not output_dir.exists():
#         output_dir.mkdir()
#     nlp.to_disk(output_dir)
#     print("Saved model to", output_dir)

    # test the saved model
#         print("Loading from", output_dir)
#         nlp2 = spacy.load(output_dir)
#         for text, _ in TRAIN_DATA:
#             doc = nlp2(text)
#             print("Entities", [(ent.text, ent.label_) for ent in doc.ents])
#             print("Tokens", [(t.text, t.ent_type_, t.ent_iob) for t in doc])
```

```python
dataframe_to_populate = pd.DataFrame()
for i in dataframe.filename.unique():
    print(i)
    dataframe_to_populate = dataframe_to_populate.append(pd.DataFrame({'filename':[i],'fname':[''],'sname':[''],'sdate':[''],'temp':[''],'edate':[''],'rent':[''],'dur':[''],'notice':['']}),ignore_index=True)

#     re.sub('MRS|Mrs|MR|Mr','')
#     print(dataframe[dataframe.filename == i]['fname'].unique())
#     print(dataframe[dataframe.filename == i]['sname'].unique())
#     print(dataframe[dataframe.filename == i]['sdate'].unique())
#     print(dataframe[dataframe.filename == i]['edate'].unique())
#     print(dataframe[dataframe.filename == i]['rent'].unique())
#     print(dataframe[dataframe.filename == i]['dur'].unique())
#     print(dataframe[dataframe.filename == i]['notice'].unique())
    
    for values in dataframe[dataframe.filename == i]['fname'].unique():
        if dataframe_to_populate[dataframe_to_populate.filename == i]['fname'].values[0] == '' :
            if values != 'x':
                dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'fname'] = re.sub('MRS|Mrs|MR|Mr|Sri|Smt','',values)
                
    for values in dataframe[dataframe.filename == i]['sname'].unique():
        if dataframe_to_populate[dataframe_to_populate.filename == i]['sname'].values[0] == '' :
            if values != 'x':
                dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'sname'] = re.sub('MRS|Mrs|MR|Mr|Sri|Smt','',values)
                
    for values in dataframe[dataframe.filename == i]['sdate'].unique():
        if dataframe_to_populate[dataframe_to_populate.filename == i]['sdate'].values[0] == '' :
            if values != 'x':
                try:
                    values = re.sub('day of|day|date|jalan','',values.lower())
                    dte = pd.to_datetime(values)
                    dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'temp'] = dte
                    dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'sdate'] = str(dte.day)+'.'+str(dte.month)+'.'+str(dte.year)
                except:
                    dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'sdate'] = values
    
    
    
    for values in dataframe[dataframe.filename == i]['rent'].unique():
        if dataframe_to_populate[dataframe_to_populate.filename == i]['rent'].values[0] == '' :
            if values != 'x':
                values = re.sub('/-','',values)
                values = re.sub('RM|RS|S','',values)
                if len(values.split()) > 2:
                    values = ''.join(values.split()[:2])
                if len(values.split())>1:
                    if len(values.split()[0])>2:
                        values = values.split()[0]
                dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'rent'] = values
                
    for values in dataframe[dataframe.filename == i]['dur'].unique():
        if dataframe_to_populate[dataframe_to_populate.filename == i]['dur'].values[0] == '' :
            if values != 'x':
                values = re.sub('months|month','',values.lower())
                try:
                    if len(values)>2:
                        values = w2n.word_to_num(values)
                except:
                    pass
                dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'dur'] = values
                
                
    for values in dataframe[dataframe.filename == i]['edate'].unique():
        if dataframe_to_populate[dataframe_to_populate.filename == i]['edate'].values[0] == '' :
            if values != 'x':
                try:
                    values = re.sub('day of|day|date','',values.lower())
                    dte = pd.to_datetime(values)
#                     dte = dte+relativedelta(months=+dataframe_to_populate[dataframe_to_populate['filename']==i]['dur'].values[0])
                    dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'edate'] = str(dte.day)+'.'+str(dte.month)+'.'+str(dte.year)
                except:
                    dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'edate'] = values
                
    for values in dataframe[dataframe.filename == i]['notice'].unique():
        if dataframe_to_populate[dataframe_to_populate.filename == i]['notice'].values[0] == '' :
            if values != 'x':
                values = re.sub('months','',values.lower())
                try:
                    if len(values)>2:
                        values = w2n.word_to_num(values)
                except:
                    pass
                
                try:
                    values = int(values) * 30
                except:
                    pass
                dataframe_to_populate.loc[(dataframe_to_populate['filename']==i),'notice'] = values
    
```
